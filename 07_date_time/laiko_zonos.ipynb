```python
from datetime import datetime
from zoneinfo import ZoneInfo, ZoneInfoNotFoundError # Importuojame laiko zonos klasę

ivestas_laikas = input("Iveskite laika formatu HH:MM(:SS), sekundes nera butinos: ")
formatas = "%H:%M"

# Tikriname, ar ivestas laikas yra tinkamo formato
try:
    if len(ivestas_laikas) > 5:
        formatas = "%H:%M:%S"
    laikas = datetime.strptime(ivestas_laikas, formatas)
except ValueError:
    print("Netinkamas laiko formatas. Prašome įvesti tinkamą laiką.") # Jei netinkamas laiko formatas
    exit(1) # programa išeina

vartotojo_laiko_zona = input("Iveskite laiko zona pvz.: \"America/Barbados\": ") # Iveskite laiko zoną

# Tikriname, ar ivesta laiko zona yra tinkama
try:
    sydney = laikas.replace(tzinfo=ZoneInfo(vartotojo_laiko_zona)).astimezone(ZoneInfo("Australia/Sydney"))
    dubai = laikas.replace(tzinfo=ZoneInfo(vartotojo_laiko_zona)).astimezone(ZoneInfo("Asia/Dubai"))
    vilnius = laikas.replace(tzinfo=ZoneInfo(vartotojo_laiko_zona)).astimezone(ZoneInfo("Europe/Vilnius"))
    london = laikas.replace(tzinfo=ZoneInfo(vartotojo_laiko_zona)).astimezone(ZoneInfo("Europe/London"))
    new_york = laikas.replace(tzinfo=ZoneInfo(vartotojo_laiko_zona)).astimezone(ZoneInfo("America/New_York"))
    los_angeles = laikas.replace(tzinfo=ZoneInfo(vartotojo_laiko_zona)).astimezone(ZoneInfo("America/Los_Angeles"))
except ZoneInfoNotFoundError:
    print("Netinkama laiko zona. Prašome įvesti tinkamą laiko zoną.") # Jei netinkama laiko zona
    exit(1)

print(f"""Jūsų laikas: {laikas.time()}, {vartotojo_laiko_zona}
Sydney laikas: {sydney.time()}, Australia/Sydney
Dubajaus laikas: {dubai.time()}, Asia/Dubai
Vilniaus laikas: {vilnius.time()}, Europe/Vilnius
Londono laikas: {london.time()}, Europe/London
New Yorko laikas: {new_york.time()}, America/New York
Los Angeles laikas: {los_angeles.time()}, America/Los_Angeles""")
```

```python
from datetime import datetime # datetime is a class in the datetime module
from zoneinfo import ZoneInfo # ZoneInfo is a class in the zoneinfo module

def get_formatted_time(ivestas_laikas, formatas): # formatas is a string
    return datetime.strptime(ivestas_laikas, formatas) # strptime is a method of the datetime class
    # .astimezone is a method of the datetime class
def convert_and_print_timezone(laikas, vartotojo_laiko_zona, target_timezone): # target_timezone is a string
    converted_time = laikas.replace(tzinfo=ZoneInfo(vartotojo_laiko_zona)).astimezone(ZoneInfo(target_timezone)) # replace is a method of the datetime class, astimezone is a method of the datetime class
    print(f"{target_timezone.split('/')[-1]} laikas: {converted_time.time()}, {target_timezone}") # split is a method of the string class

def main():
    ivestas_laikas = input("Iveskite laika formatu HH:MM(:SS), sekundes nera butinos: ")
    formatas = "%H:%M" if len(ivestas_laikas) <= 5 else "%H:%M:%S" # len is a method of the string class

    vartotojo_laiko_zona = input("Iveskite laiko zona pvz.: \"America/Barbados\": ")

    laikas = get_formatted_time(ivestas_laikas, formatas)

    print(f"Jusu laikas: {laikas.time()}, {vartotojo_laiko_zona}") 
    
    timezones = ["Australia/Sydney", "Asia/Dubai", "Europe/Vilnius", "Europe/London", "America/New_York", "America/Los_Angeles"] 

    for timezone in timezones:
        convert_and_print_timezone(laikas, vartotojo_laiko_zona, timezone) # passing arguments to a function

if __name__ == "__main__":
    main()
```
